```python
data = data.dropna(subset=["Suggested Job Role"])
```

```python
job_role_mapping = {
    "Network Security Administrator": "Network & Security",
    "Network Security Engineer": "Network & Security",
    "Network Engineer": "Network & Security",
    "Systems Security Administrator": "Network & Security",
    "Information Security Analyst": "Network & Security",
    "Information Technology Auditor": "Network & Security",
    "Project Manager": "Management & Bussiness",
    "Information Technology Manager": "Management & Bussiness",
    "Database Administrator": "Database & Data Management",
    "Portal Administrator": "Database & Data Management",
    "Database Developer": "Database & Data Management",
    "Database Manager": "Database & Data Management",
    "Data Architect": "Database & Data Management",
    "Software Engineer": "Software Development & Engineering",
    "Software Developer": "Software Development & Engineering",
    "Software Systems Engineer": "Software Development & Engineering",
    "Solutions Architect": "Software Development & Engineering",
    "Applications Developer": "Software Development & Engineering",
    "Programmer Analyst": "Software Development & Engineering",
    "Web Developer": "Software Development & Engineering",
    "UX Designer": "UI/UX & Design",
    "Design & UX": "UI/UX & Design",
    "CRM Business Analyst": "Data Science & Analytics",
    "Business Systems Analyst": "Data Science & Analytics",
    "Business Intelligence Analyst": "Data Science & Analytics",
    "E-Commerce Analyst": "Data Science & Analytics",
    "CRM Technical Developer": "CRM & Technical Support",
    "Technical Support": "CRM & Technical Support",
    "Technical Services/Help Desk/Tech Support": "CRM & Technical Support",
    "Technical Engineer": "CRM & Technical Support",
    "Software Quality Assurance (QA) / Testing": "Software Testing & Quality Assurance",
    "Quality Assurance Associate": "Software Testing & Quality Assurance",
    "Mobile Applications Developer": "Software Development & Engineering"
}
```

```python
data["Broad Job Category"] = data["Suggested Job Role"].map(job_role_mapping)
```

```python
data = data.dropna(subset=["Broad Job Category"])  # Drop unmapped roles
```

```python
data["Broad Job Category"].value_counts()
```

```python
data["Interested Type of Books"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode categorical columns
label_encoders = {}
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
data.head()
```

```python
data["Broad Job Category"].value_counts()
```

```python
from sklearn.utils import resample
```

```python
# Balance the dataset using oversampling
max_class_size = data["Broad Job Category"].value_counts().max()
df_balanced = data.groupby("Broad Job Category", group_keys=False).apply(
    lambda x: resample(x, replace=True, n_samples=max_class_size, random_state=42)
)
```

```python
# Split balanced dataset into features and target
X_bal = df_balanced.drop(columns=["Suggested Job Role", "Broad Job Category"])
y_bal = df_balanced["Broad Job Category"]
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
```

```python
#feature selection
rfe = RFE(estimator=RandomForestClassifier(), n_features_to_select=10)
X_rfe = rfe.fit_transform(X_bal, y_bal)
```

```python
print(rfe.support_)
```

```python
print(rfe.ranking_)
```

```python
#printing feature selected by rfe
for feature in rfe.support_.nonzero():
  print(data.columns[feature])
```

```python
data["Interested Type of Books"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split into training and testing sets/
X_train, X_test, y_train, y_test = train_test_split(X_rfe, y_bal, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Train model with selected features
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Evaluate model before optimization
accuracy_before = accuracy_score(y_test, y_pred)
print("Accuracy Before Optimization:", accuracy_before)
print("Classification Report Before Optimization:\n", classification_report(y_test, y_pred))
```

```python
param_grid = {
    "n_estimators": [50, 100, 150,200],
    "max_depth": [10, 20, 15,None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring="accuracy", n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
# Best parameters
tuned_params = grid_search.best_params_
tuned_params
```

```python
# Train optimized model
optimized_model = RandomForestClassifier(**tuned_params, random_state=42)
```

```python
optimized_model.fit(X_train, y_train)
```

```python
y_pred_optimized = optimized_model.predict(X_test)
```

```python
# Evaluate optimized model
accuracy_optimized = accuracy_score(y_test, y_pred_optimized)
print("Optimized Accuracy:", accuracy_optimized)
print("Optimized Classification Report:\n", classification_report(y_test, y_pred_optimized))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode categorical columns
label_encoders = {}
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
from sklearn.utils import resample
```

```python
# Balance the dataset using oversampling
max_class_size = data["Broad Job Category"].value_counts().max()
df_balanced = data.groupby("Broad Job Category", group_keys=False).apply(
    lambda x: resample(x, replace=True, n_samples=max_class_size, random_state=42)
)
```

```python
# Split balanced dataset into features and target
X_bal = df_balanced.drop(columns=["Suggested Job Role", "Broad Job Category"])
y_bal = df_balanced["Broad Job Category"]
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
```

```python
#feature selection
rfe = RFE(estimator=RandomForestClassifier(), n_features_to_select=10)
X_rfe = rfe.fit_transform(X_bal, y_bal)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split into training and testing sets/
X_train, X_test, y_train, y_test = train_test_split(X_rfe, y_bal, test_size=0.2, random_state=42)
```

```python
# Train model with selected features
model = RandomForestClassifier(n_estimators=200,max_depth=20 ,random_state=42)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Evaluate model before optimization
accuracy_before = accuracy_score(y_test, y_pred)
print("Accuracy Before Optimization:", accuracy_before)
print("Classification Report Before Optimization:\n", classification_report(y_test, y_pred))
```

```python
#now we will do cross validation 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#logistic regression
lr_cr_val=cross_val_score(LogisticRegression(random_state=10),X_train, y_train)
print(lr_cr_val)
print(np.mean(lr_cr_val))
```

```python
#svm
svm_cr_val=cross_val_score(SVC(C=1,kernel="rbf",gamma=10),X_train, y_train)
print(svm_cr_val)
```

```python
#random fores
rfc_cr_val=cross_val_score(RandomForestClassifier(random_state=42),X_train, y_train)
print(rfc_cr_val)
print(np.mean(rfc_cr_val))
```

```python
#decision tree
dt_cr_val=cross_val_score(DecisionTreeClassifier(random_state=42,max_depth=20),X_train, y_train)
print(dt_cr_val)
print(np.mean(dt_cr_val))
```

```python
#AdaBoost
ada_cr_val=cross_val_score(AdaBoostClassifier(random_state=42,algorithm='SAMME',learning_rate=2,n_estimators=100),X_train, y_train)
print(ada_cr_val)
print(np.mean(ada_cr_val))
```

```python
#gradientboosting
xgboost_cr_val=cross_val_score(GradientBoostingClassifier(random_state=42,learning_rate=2,n_estimators=200),X_train, y_train)
print(xgboost_cr_val)
print(np.mean(xgboost_cr_val))
```

```python
#xgboost
xg=cross_val_score(XGBClassifier(random_state=42,n_estimators=200,learning_rate=1),X_train, y_train)
print(xg)
print(np.mean(xg))
```

```python
#hyperparameter tunning for SVM
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    'C': np.logspace(-3,1, 2, 10),                    
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'], 
    'degree': [2, 3, 4, 5],                         
    'gamma': ['scale', 'auto'] + list(np.logspace(-4, 1, 6)),  
    'shrinking': [True, False],                    
    'probability': [True],                         
    'tol': [1e-3, 1e-4, 1e-5]                       
}
```

```python
svc = SVC(random_state=42)
```

```python
random_search = RandomizedSearchCV(
    estimator=svc,
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
random_search.fit(X_train, y_train)
```
